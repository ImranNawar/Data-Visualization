**Matplotlib can work with Pandas DataFrames, but the integration is not as seamless or high-level as Seaborn's.**

```python
import matplotlib.pyplot as plt
import pandas as pd

data = tips

# Create a scatter plot
plt.scatter(data['total_bill'], data['tip'], c=data['day'].astype('category').cat.codes)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter Plot of Tips vs Total Bill')
plt.colorbar(label='Day')
plt.show()
```

`cat.codes` converts categorical values in a column (like labels or names) into corresponding numerical codes, where each unique category is assigned a unique integer.

### Visualization of Numeric Data

**Bar Chart**

```python
plt.bar(['Product A', 'Product B', 'Product C'], [10, 20, 15])
plt.title('Sales by Product')
plt.show()
```

**Line Plot**

```python
""" Simple Line plot """

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title('Stock Prices Over Time')
plt.show()
```

```python
"""Customized Line Plot"""

# Data for the plot
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
y2 = [8, 15, 30, 25]

# Create the plot
fig=plt.figure(figsize=(8, 3))
fig.add_subplot(1, 2, 1)
plt.plot(x, y, marker='+', linestyle='--', color='y', label='Stock A')
fig.add_subplot(1, 2, 2)
plt.plot(x, y2, marker='x', linestyle='-', color='r', label='Stock B')

# Add title and labels
plt.title('Stock Prices Over Time', fontsize=8)
plt.xlabel('Time (Days)', fontsize=8)
plt.ylabel('Stock Price (USD)', fontsize=8)

# Add a grid
plt.grid(True)

# Add a legend
plt.legend()

plt.show()
```

**Scatter Plot**

```python
heights = [1.5, 1.8, 1.6, 2.0, 1.7, 1.9, 1.75, 1.65]
weights = [65, 80, 70, 90, 75, 85, 78, 68]

plt.scatter(heights, weights, color="g", marker='o')
plt.title('Height vs Weight', fontsize=14)
plt.xlabel('Height (m)', fontsize=12)
plt.ylabel('Weight (kg)', fontsize=12)

plt.show()
```

**Histogram**

```python
# Data for exam scores
scores = [70, 85, 90, 75, 80, 90, 85, 80, 88, 92, 77, 84, 79]

plt.hist(scores, bins = 7, color='skyblue', edgecolor='black')

plt.title('Distribution of Exam Scores', fontsize=14)
plt.xlabel('Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.show()
```

### Visualizing of Categorical Data

**Bar Plot**

```python
# Data for age groups
age_groups = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']
number_of_people = [50, 70, 60, 40, 30, 20]

plt.bar(age_groups, number_of_people, color='skyblue', edgecolor='black')

plt.title('Number of People in Each Age Group', fontsize=14)
plt.xlabel('Age Group')
plt.ylabel('Number of People')

plt.show()
```

**Count Plot**

```python
import pandas as pd

data = {'Brand': ['Toyota', 'Honda', 'Ford', 'Toyota', 'Ford', 'Honda', 'Toyota', 'Ford', 'Toyota']}
df = pd.DataFrame(data)

# Create the count plot
sns.countplot(x='Brand', data=df, palette='viridis')

# Add title and axis labels
plt.title('Number of Cars by Brand')
plt.xlabel('Car Brand')
plt.ylabel('Count')

plt.show()
```

**Pie Chart**

```python
# Data for market share
companies = ['Company A', 'Company B', 'Company C', 'Company D']
market_share = [35, 25, 20, 20]

plt.pie(market_share, labels=companies, autopct='%1.2f%%' ,colors=['lightblue', 'lightgreen', 'lightcoral', 'lightskyblue'])

# Add title
plt.title('Market Share of Different Companies', fontsize=14)
plt.show()
```
